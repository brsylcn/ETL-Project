```python
#Import Dependancies
import pandas as pd
```

```python
csv_file = "./Final_VGData.csv"
FVGData_df = pd.read_csv(csv_file)
FVGData_df
```

```python
FVGData_clean_df = FVGData_df.dropna().drop_duplicates()
FVGData_clean_df = FVGData_clean_df.reset_index()
FVGData_clean_df
```

```python
FVGData_clean_df.dtypes
```

```python
FVGData_clean_df['Year_of_Release'] = FVGData_clean_df['Year_of_Release'].astype(int)
FVGData_clean_df
```

```python
FVGData_clean_df = FVGData_clean_df.drop(['index','ID_key'], axis=1)
FVGData_clean_df = FVGData_clean_df.reset_index()
FVGData_clean_df = FVGData_clean_df.rename(columns={"index": "game_id"})
FVGData_clean_df = FVGData_clean_df.set_index("game_id")
FVGData_clean_df
```

```python
platform = FVGData_clean_df["Platform"]
platform = platform.drop_duplicates().reset_index()
platform = platform.drop(['game_id'], axis=1)
platform = platform.reset_index()
platform = platform.rename(columns={"index": "platform_id"})
#platform = platform.set_index("platform_id")
platform
```

```python
genre = FVGData_clean_df["Genre"]
genre = genre.drop_duplicates().reset_index()
genre = genre.drop(['game_id'], axis=1)
genre = genre.reset_index()
genre = genre.rename(columns={"index": "genre_id"})
#genre = genre.set_index("genre_id")
genre
```

```python
publisher = FVGData_clean_df["Publisher"]
publisher = publisher.drop_duplicates().reset_index()
publisher = publisher.drop(['game_id'], axis=1)
publisher = publisher.reset_index()
publisher = publisher.rename(columns={"index": "publisher_id"})
#publisher = publisher.set_index("publisher_id")
publisher
```

```python
release_year = FVGData_clean_df["Year_of_Release"]
release_year = release_year.sort_values().astype(int)
release_year = release_year.drop_duplicates().reset_index()
release_year = release_year.drop(['game_id'], axis=1)
release_year = release_year.reset_index()
release_year = release_year.rename(columns={"index": "rel_year_id"})
#release_year = release_year.set_index("rel_year_id")
release_year
```

```python
sales = FVGData_clean_df[['NA_Sales','EU_Sales','JP_Sales','Global_Sales']]
sales
```

```python
FVGData_wPlatID = pd.merge(FVGData_clean_df,platform,on="Platform", how="left")

FVGData_wPlatID.head(40)
```

```python
FVGData_wGenreID = pd.merge(FVGData_wPlatID,genre,on="Genre", how="left")

FVGData_wGenreID.head(40)
```

```python
FVGData_wPubID = pd.merge(FVGData_wGenreID,publisher,on="Publisher", how="left")

FVGData_wPubID.head(40)
```

```python
FVGData_wYearID = pd.merge(FVGData_wPubID,release_year,on="Year_of_Release", how="left")

FVGData_wYearID.head(40)
```

```python
FVGData_wYearID.columns
```

```python
games_main = FVGData_wYearID[['Name', 'platform_id',
       'genre_id', 'publisher_id', 'rel_year_id']]

games_main
```
