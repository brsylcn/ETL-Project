```python
FVGData_df = Final_VGData
FVGData_df
```

```python
FVGData_clean_df = FVGData_df.dropna().drop_duplicates()
FVGData_clean_df = FVGData_clean_df.reset_index()
FVGData_clean_df
```

```python
FVGData_clean_df.dtypes
```

```python
FVGData_clean_df['Year_of_Release'] = FVGData_clean_df['Year_of_Release'].astype(int)
FVGData_clean_df
```

```python
FVGData_clean_df = FVGData_clean_df.drop(['index','ID_key'], axis=1)
FVGData_clean_df = FVGData_clean_df.reset_index()
FVGData_clean_df = FVGData_clean_df.rename(columns={"index": "game_id"})
FVGData_clean_df = FVGData_clean_df.set_index("game_id")
FVGData_clean_df.rename(columns={"Name": "game_name","Platform":"platform_name","Year_of_Release":"release_year","Genre":"genre","Publisher":"publisher_name","NA_Sales":"na_sales","EU_Sales":"eu_sales","JP_Sales":"jp_sales","Other_Sales":"othr_sales","Global_Sales":"gbl_sales"}, inplace=True)
FVGData_clean_df
```

```python
string = FVGData_clean_df.to_html(index=False, classes ="table")
with open('table.html', 'w', encoding='utf-8') as html_table:
    html_table.write(string)
```

```python
platform = FVGData_clean_df["platform_name"]
platform = platform.drop_duplicates().reset_index()
platform = platform.drop(['game_id'], axis=1)
platform = platform.reset_index()
platform = platform.rename(columns={"index": "platform_id"})
#platform = platform.set_index("platform_id")
platform
```

```python
genre = FVGData_clean_df["genre"]
genre = genre.drop_duplicates().reset_index()
genre = genre.drop(['game_id'], axis=1)
genre = genre.reset_index()
genre = genre.rename(columns={"index": "genre_id"})
#genre = genre.set_index("genre_id")
genre
```

```python
publisher = FVGData_clean_df["publisher_name"]
publisher = publisher.drop_duplicates().reset_index()
publisher = publisher.drop(['game_id'], axis=1)
publisher = publisher.reset_index()
publisher = publisher.rename(columns={"index": "publisher_id"})
#publisher = publisher.set_index("publisher_id")
publisher
```

```python
release_year = FVGData_clean_df["release_year"]
release_year = release_year.sort_values().astype(int)
release_year = release_year.drop_duplicates().reset_index()
release_year = release_year.drop(['game_id'], axis=1)
release_year = release_year.reset_index()
release_year = release_year.rename(columns={"index": "rel_year_id"})
#release_year = release_year.set_index("rel_year_id")
release_year
```

```python
sales = FVGData_clean_df[['na_sales','eu_sales','jp_sales','othr_sales','gbl_sales']]
sales
```

```python
FVGData_wPlatID = pd.merge(FVGData_clean_df,platform,on="platform_name", how="left")
FVGData_wPlatID.head(40)
```

```python
FVGData_wGenreID = pd.merge(FVGData_wPlatID,genre,on="genre", how="left")
FVGData_wGenreID.head(40)
```

```python
FVGData_wPubID = pd.merge(FVGData_wGenreID,publisher,on="publisher_name", how="left")
FVGData_wPubID.head(40)
```

```python
FVGData_wYearID = pd.merge(FVGData_wPubID,release_year,on="release_year", how="left")
FVGData_wYearID.head(40)
```

```python
FVGData_wYearID.columns
```

```python
games_main = FVGData_wYearID[['game_name', 'platform_id',
       'genre_id', 'publisher_id', 'rel_year_id']]
games_main
```

```python
games_main = games_main.reset_index()
games_main = games_main.rename(columns={"index": "game_id"})
games_main = games_main.set_index("game_id")
games_main
```

# Load

### Connect to local database

```python
#Make sure you create a DB in PGAdmin called ETL_db and use the appropriate credentials for your local postgres DB
rds_connection_string = "postgres:postGRES12@localhost:5432/ETL_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

### Push data to DB

```python
sales.to_sql(name='sales', con=engine, if_exists='replace')
```

```python
release_year.to_sql(name='release_year', con=engine, if_exists='append', index=False)
```

```python
genre.to_sql(name='genre', con=engine, if_exists='append', index=False)
```

```python
publisher.to_sql(name='publisher', con=engine, if_exists='append', index=False)
```

```python
platform.to_sql(name='platform', con=engine, if_exists='append', index=False)
```

```python
games_main.to_sql(name='games_main', con=engine, if_exists='append', index=True)
```
