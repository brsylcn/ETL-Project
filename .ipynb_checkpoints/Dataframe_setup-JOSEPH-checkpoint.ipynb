```python
# Import dependencies for data transformation, analysis and plotting

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
import random
import seaborn as sns
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode, iplot
```

```python
# Import PS4 dataset
PS4_data = pd.read_csv("PS4_GamesSales.csv", encoding = "ISO-8859-1")   
PS4_data.head()
```

```python
# Add Platform column to PS4
Platform = "PS4"
PS4_data["Platform"] = Platform
PS4_data.head()
```

```python
# Import Xbox dataset

XboxOne_data = pd.read_csv("XboxOne_GameSales.csv", encoding = "ISO-8859-1")    
XboxOne_data.head()
```

```python
# Add platform column to Xbox

Platform = "XOne"
XboxOne_data["Platform"] = Platform
XboxOne_data.head()
```

```python
# Import Video Game Sales dataset

Video_Games_Sales = pd.read_csv("Video_Games_Sales_as_at_22_Dec_2016.csv", encoding = "ISO-8859-1")    #('../Resources/trip.csv')
Video_Games_Sales.head()
```

```python
# Add ID key to Video Game Sales dataframe

ID_key = Video_Games_Sales["Name"] + Video_Games_Sales["Platform"]
Video_Games_Sales["ID_key"] = ID_key
Video_Games_Sales.head()
```

```python
# Reorganize and remove unnecessary columns from Video Game Sales dataframe

Final_VGSales = Video_Games_Sales[['ID_key','Name', 'Platform', 'Year_of_Release', 'Genre', 'Publisher', 'NA_Sales',
       'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]
Final_VGSales.head()
```

```python
# Add ID key to XboxOne dataframe

ID_key = XboxOne_data["Game"] + XboxOne_data["Platform"]
XboxOne_data["ID_key"] = ID_key
XboxOne_data.head()
```

```python
# Add ID key to PS4 dataframe

ID_key = PS4_data["Game"] + PS4_data["Platform"]
PS4_data["ID_key"] = ID_key
PS4_data.head()
```

```python
# Reorganize PS4 columns to match Video Game Sales columns

organized_PS4 = PS4_data[['ID_key', 'Game','Platform', 'Year', 'Genre', 'Publisher', 'North America', 'Europe',
       'Japan', 'Rest of World', 'Global']]
organized_PS4.head()
```

```python
# Rename PS4 columns to match Video Game Sales columns

Final_PS4 = organized_PS4.rename(columns={"Game":"Name", "Year":"Year_of_Release", "North America":"NA_Sales","Europe":"EU_Sales","Japan":"JP_Sales","Rest of World":"Other_Sales","Global":"Global_Sales"})
Final_PS4.head()
```

```python
# Reorganize Xbox columns to match Video Game Sales columns

organized_Xbox = XboxOne_data[['ID_key', 'Game','Platform', 'Year', 'Genre', 'Publisher', 'North America', 'Europe',
       'Japan', 'Rest of World', 'Global']]
organized_Xbox.head()
```

```python
# Rename Xbox columns to match Video Game Sales columns

Final_XOne = organized_Xbox.rename(columns={"Game":"Name", "Year":"Year_of_Release", "North America":"NA_Sales","Europe":"EU_Sales","Japan":"JP_Sales","Rest of World":"Other_Sales","Global":"Global_Sales"})
Final_XOne.head()
```

```python
Final_VGSales
```

```python
Sales_plus_PS4 = Final_VGSales.append(Final_PS4)
Sales_plus_PS4
```

```python
Final_VGData = Sales_plus_PS4.append(Final_XOne)
Final_VGData
```

```python
# Regional and Rest-of-world Sales by Publisher

publisher_NA = Final_VGData.groupby("Publisher")["NA_Sales"].sum()
publisher_EU = Final_VGData.groupby("Publisher")["EU_Sales"].sum()
publisher_JP = Final_VGData.groupby("Publisher")["JP_Sales"].sum()
publisher_Other = Final_VGData.groupby("Publisher")["Other_Sales"].sum()


# Regional and Rest-of-world Sales by Publisher

platform_NA = Final_VGData.groupby("Platform")["NA_Sales"].sum()
platform_EU = Final_VGData.groupby("Platform")["EU_Sales"].sum()
platform_JP = Final_VGData.groupby("Platform")["JP_Sales"].sum()
platform_Other = Final_VGData.groupby("Platform")["Other_Sales"].sum()
```

```python
# Global Sales by Platform Chart

fig=plt.figure(figsize=(30,20))
plt.subplots_adjust(left=0.25, wspace=0.20, hspace=0.35)
sns.set_style("white")

plt.subplot(2, 2, 1)
plt.title('Global Sales by Platform',fontdict={'fontsize':16})
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.barplot(y='Platform', x='Global_Sales', data=Final_VGData.groupby('Platform').sum().Global_Sales.sort_values(ascending=False).reset_index(),palette='YlOrRd_r');
plt.ylabel('Platform',fontdict={'fontsize':16})
plt.xlabel('Sales in millions',fontdict={'fontsize':16})
```

```python
# Global Sales by Publisher chart

fig=plt.figure(figsize=(30,20))
plt.subplots_adjust(left=0.25, wspace=0.20, hspace=0.35)
sns.set_style("white")

plt.subplot(2, 2, 1)
plt.title('Top 25 Global Sales by Publisher',fontdict={'fontsize':16})
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.barplot(y='Publisher', x='Global_Sales', data=Final_VGData.groupby('Publisher').sum().nlargest(25,'Global_Sales').Global_Sales.sort_values(ascending=False).reset_index(),palette='YlOrRd_r');
plt.ylabel('Publisher',fontdict={'fontsize':16})
plt.xlabel('Sales in millions',fontdict={'fontsize':16})
```
